```python
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
estado_civil = [0, 1, 0, 1, 0]
renda = [1000, 2000, 3500, 4000, 5000]
aprovacao = [0, 1, 0, 1, 1]

df_aprovacao = pd.DataFrame({"estado_civil": estado_civil, "renda_mensal": renda, "aprovado": aprovacao})
```

```python

X = df_aprovacao.iloc[:,:2]
y = df_aprovacao.iloc[:,-1]

clf = tree.DecisionTreeClassifier()

clf.fit(X, y)
```

```python



plt.figure(figsize=(8,6))
tree.plot_tree(clf)
```
