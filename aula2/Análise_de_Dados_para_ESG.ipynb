Seu primeiro código usando Pandas


```python
import pandas as pd

# Criando um conjunto de dados simples
dados = {
    "Nome": ["Ana", "Bruno", "Carlos", "Daniela"],
    "Idade": [21, 25, 19, 23],
    "Nota": [8.5, 7.0, 9.2, 6.8]
}

# Criando um DataFrame
df = pd.DataFrame(dados)
#df = pd.read_csv("amostra1.csv")

# Mostrando os dados
print(df)

# Média das notas
media = df["Nota"].mean()
print("Média das notas:", media)
```
