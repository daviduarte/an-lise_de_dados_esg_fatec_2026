```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Dados (chumbados)
# -----------------------------
estudo = np.array([1, 2, 3, 4, 5, 6])
frequencia = np.array([5, 6, 7, 8, 9, 9])
nota = np.array([4, 5, 6, 7, 8, 9])

# -----------------------------
# Coeficientes (ajuste manual)
# Modelo: Nota = w1*Estudo + w2*Frequencia + c
# -----------------------------
w1 = 1
w2 = 0
c = 3

# -----------------------------
# Criando grid para o plano
# -----------------------------
x = np.linspace(estudo.min(), estudo.max(), 10)
y = np.linspace(frequencia.min(), frequencia.max(), 10)
X, Y = np.meshgrid(x, y)

Z = w1 * X + w2 * Y + c

# -----------------------------
# Plot
# -----------------------------
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Pontos reais
ax.scatter(estudo, frequencia, nota)

# Plano estimado
ax.plot_surface(X, Y, Z, alpha=0.3)

ax.view_init(elev=30, azim=30)

# Labels
ax.set_xlabel('Estudo')
ax.set_ylabel('Frequência')
ax.set_zlabel('Nota')

plt.show()
```
